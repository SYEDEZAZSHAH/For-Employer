Q1. How have global temperature anomalies trended over time?

To answer this:

Data: We'll use the global_temps dataset to observe yearly anomalies.
Narrative: After analyzing the data, we'll discuss the overall trend, pointing out periods of significant increase or decrease.
Visualizations: We'll create a line graph showing the trend of global temperature anomalies over the years.
Let's begin by visualizing the trend of global temperature anomalies over the years using the "J-D" column from the global_temps dataset.

```python
import matplotlib.pyplot as plt

# Plotting the data
plt.figure(figsize=(15, 7))
plt.plot(global_temps['Year'], global_temps['J-D'], label='Yearly Anomaly', color='blue', linewidth=1.5)
plt.axhline(0, color='red', linestyle='--', linewidth=0.8)  # Add a reference line at 0 for anomaly
plt.title('Trend of Global Temperature Anomalies Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (C)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```

Narrative:

The visualization provides a clear overview of the trend of global temperature anomalies from the year 1880 up to the present. Here are some observations:

Historical Perspective: From 1880 to the early 1900s, there was variability in the anomalies, but no clear trend is observed.
Mid 20th Century: Starting around the mid-20th century, there's a noticeable upward trend, indicating a gradual increase in temperature anomalies.
Recent Decades: The rate of increase becomes more pronounced from the 1980s onwards. This period aligns with the time when concerns about global warming and climate change became more mainstream.
Anomaly Reference: The dashed red line represents the zero anomaly mark. Most of the recent years have temperature anomalies above this line, indicating that they are warmer than the reference period.
In summary, there's a clear upward trend in global temperature anomalies over the years, especially in recent decades. This trend is consistent with the broader narrative about global warming and its potential impacts on the environment and ecosystems.

The visualization provides a comprehensive view of the average temperature anomalies by decade, starting from the 1880s. Here are the key observations:

Early Decades: From the 1880s to the 1910s, the temperature anomalies fluctuated around the zero mark, indicating relatively stable temperatures with respect to the reference period.

Mid 20th Century: There's a noticeable dip in the 1960s, followed by a consistent and sharp upward trend from the 1970s onwards. This aligns with the period when industrialization and urbanization began to accelerate globally.

Recent Decades: The most recent decades, especially the 2000s and 2010s, have the highest average temperature anomalies. The 2010s, in particular, stand out as the decade with the most significant spike in temperature anomalies.

Q7 Are there any outliers in the global temperature anomaly data that need further investigation?

To answer this:

Data: We'll use statistical methods to identify outliers in the global_temps dataset.
Narrative: After identifying potential outliers, we'll discuss potential reasons for these outliers, such as volcanic eruptions or significant climatic events.
Visualizations: We'll use a boxplot or scatter plot to show anomalies over time, with outliers highlighted.
Let's start by visualizing the temperature anomalies over the years using a scatter plot and then identifying potential outliers.

```python
# Plotting the temperature anomalies with potential outliers highlighted
plt.figure(figsize=(15, 7))

# Scatter plot for anomalies
plt.scatter(global_temps['Year'], global_temps['J-D'], color='blue', s=10, label='Yearly Anomalies')

# Calculate the mean and standard deviation
mean_anomaly = global_temps['J-D'].mean()
std_anomaly = global_temps['J-D'].std()

# Highlight potential outliers (more than 2 standard deviations from the mean)
outliers = global_temps[abs(global_temps['J-D'] - mean_anomaly) > 2 * std_anomaly]
plt.scatter(outliers['Year'], outliers['J-D'], color='red', s=50, label='Outliers', edgecolors='black', zorder=5)

plt.axhline(0, color='grey', linestyle='--', linewidth=0.8)  # Add a reference line at 0 for anomaly
plt.title('Global Temperature Anomalies Over Time with Outliers Highlighted')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (C)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()

# Returning the outlier years for further discussion
outliers[['Year', 'J-D']]
```

Narrative:

The scatter plot provides a visual representation of global temperature anomalies over time, with potential outliers highlighted in red.

Recent Outliers: The potential outliers, which are points more than 2 standard deviations from the mean, all fall within the recent years (2015-2022). These anomalies are significantly higher than the average, indicating exceptionally warmer years.

Consistent Trend: The recent years being flagged as outliers align with the consistent upward trend we observed in previous analyses. This emphasizes the rapid acceleration of global warming in recent times.

Reasons for Outliers: These outliers, particularly in the context of the last decade, could be attributed to a combination of factors. These include anthropogenic influences (like increased greenhouse gas emissions), natural variability, or significant climatic events (like El Nio).

The table below the graph lists the years identified as outliers and their corresponding temperature anomalies.

In summary, while the recent years have been identified as outliers due to their pronounced temperature anomalies, they also reflect the broader narrative of accelerated global warming.

Q10 How do the temperature anomalies of the first half of the dataset compare with the second half?

To answer this:

Data: We'll split the global_temps dataset into two halves and compare the average anomalies.
Narrative: After analyzing the data, we'll discuss the implications of the changes observed between the two halves.
Visualizations: We'll create comparative bar graphs or visual averages for each half.
Let's start by splitting the dataset into two equal halves and calculating the average temperature anomalies for each half.

```python
# Splitting the dataset into two equal halves
midpoint = len(global_temps) // 2
first_half = global_temps.iloc[:midpoint]
second_half = global_temps.iloc[midpoint:]

# Calculating the average anomalies for each half
first_half_avg = first_half['J-D'].mean()
second_half_avg = second_half['J-D'].mean()

# Plotting the average anomalies for both halves
plt.figure(figsize=(10, 6))
plt.bar(['First Half', 'Second Half'], [first_half_avg, second_half_avg], color=['cyan', 'magenta'], edgecolor='black')
plt.axhline(0, color='grey', linestyle='--', linewidth=0.8)  # Add a reference line at 0 for anomaly
plt.title('Comparison of Average Temperature Anomalies Between Two Halves')
plt.ylabel('Average Temperature Anomaly (C)')
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```

Narrative:

The bar graph provides a clear comparison of average temperature anomalies between the first and second halves of the dataset. Here are the main observations:

First Half: The first half of the dataset, representing the earlier years, has an average temperature anomaly that's very close to zero. This indicates relatively stable temperatures with respect to the reference period during this period.

Second Half: The second half, representing more recent years, exhibits a significantly higher average temperature anomaly. This is indicative of the rapid warming trend observed in the latter part of the dataset.

Implications: The stark difference between the two halves underscores the accelerated pace of global warming in recent times. While natural variability plays a role, the influence of human activities, such as greenhouse gas emissions, deforestation, and urbanization, cannot be ignored.

In summary, the comparison between the two halves of the dataset emphasizes the changing climate dynamics and the urgent need for mitigation and adaptation strategies.
